```python
n=10;
def appen2(x1,x2,x3,x4,x5,x6,x7,x8,x9,y):
    trainingdata.append([x1,x2,x3,x4,x5,x6,x7,x8,x9])
    traininglabel.append([y])
    
def appen(x1,x2,x3,x4,x5,x6,x7,x8,x9,x10):
    
                
    S= max(x1+x2+x3+x4+x5+x6+x7+x8+x9+x10,1.0);
                                
                                
    appen2(x2,x3,x4,x5,x6,x7,x8,x9,x10,S)
    appen2(x1,x3,x4,x5,x6,x7,x8,x9,x10,S)
    appen2(x1,x2,x4,x5,x6,x7,x8,x9,x10,S)
    appen2(x1,x2,x3,x5,x6,x7,x8,x9,x10,S)
    appen2(x1,x2,x3,x4,x6,x7,x8,x9,x10,S)
    appen2(x1,x2,x3,x4,x5,x7,x8,x9,x10,S)
    appen2(x1,x2,x3,x4,x5,x6,x8,x9,x10,S)
    appen2(x1,x2,x3,x4,x5,x6,x7,x9,x10,S)
    appen2(x1,x2,x3,x4,x5,x6,x7,x8,x10,S)
    appen2(x1,x2,x3,x4,x5,x6,x7,x8,x9,S)
    
def readdata():
    devided=8
    for i1 in range(devided+1):
        for i2 in range(i1+1):
            for i3 in range(i2+1):
                for i4 in range(i3+1):
                    for i5 in range(i4+1):
                        for i6 in range(i5+1):
                            for i7 in range(i6+1):
                                for i8 in range(i7+1):
                                    for i9 in range(i8+1):
                                        for i10 in range(i9+1):
                                            appen(i1/devided,i2/devided,i3/devided,i4/devided,i5/devided,
                                                  i6/devided,i7/devided,i8/devided,i9/devided,i10/devided);
                            

trainingdata=[]
traininglabel=[]
S=1.0
readdata();
```

```python
trainingdata=np.array(trainingdata)
traininglabel=np.array(traininglabel)
```

```python
#print(trainingdata)

import tensorflow as tf
#导入TensorFlow工具包并简称为tf
 
from numpy.random import RandomState
#导入numpy工具包，生成模拟数据集


w1 = tf.Variable(tf.truncated_normal([9,100],stddev=0.1))
w2 = tf.Variable(tf.truncated_normal([100,100],stddev=0.1))
w3 = tf.Variable(tf.truncated_normal([100,100],stddev=0.1))
w4 = tf.Variable(tf.truncated_normal([100,1],stddev=0.1))

biases1 = tf.Variable(tf.zeros([1,100])+0.01)
biases2 = tf.Variable(tf.zeros([1,100])+0.01)
biases3 = tf.Variable(tf.zeros([1,100])+0.01)
biases4 = tf.Variable(tf.zeros([1,1])+0.01)


lr = tf.Variable(0.01,dtype=tf.float32)
lr2 = tf.Variable(0.1,dtype=tf.float32)
#分别定义一二层和二三层之间的网络参数，标准差为1，随机产生的数保持一致

with tf.name_scope('input'):
    x = tf.placeholder(tf.float32,shape=(None,9),name='x-input')
    y_ = tf.placeholder(tf.float32,shape=(None,1),name='y-input')
#输入为两个维度，即两个特征，输出为一个标签,声明数据类型float32，None即一个batch大小
#y_是真实的标签

x1 = tf.nn.relu(tf.matmul(x,w1)+biases1)
x2 = tf.nn.relu(tf.matmul(x1,w2)+biases2)
x3 =  tf.nn.relu(tf.matmul(x2,w3)+biases3)
y =  tf.nn.relu(tf.matmul(x3,w4)+biases4)
#定义神经网络前向传播过程


X=trainingdata;
Y=traininglabel;


batch_size = n
dataset_size = len(X)
#定义训练数据batch的大小

loss=tf.where(tf.greater(tf.reduce_sum(y),tf.reduce_sum(y_)*9.0/n),  
                             tf.where(
                                 tf.greater(tf.reduce_sum(y),tf.reduce_sum(y_)*9.55/n),
                                tf.square(tf.reduce_sum(y)-tf.reduce_sum(y_)*9.55/n)/25,
                                 (0.0*tf.reduce_sum(y_))
                                     )  , 
                             tf.square(tf.reduce_sum(y_)*9.0/n-tf.reduce_sum(y))/25
                            )

train_step=tf.train.GradientDescentOptimizer(lr2).minimize(loss)

#loss3=tf.reduce_mean(tf.square(tf.reduce_sum(y)-tf.reduce_sum(y_)*6.25/7))
#loss=tf.reduce_mean(tf.where(tf.greater(y_up,y),  
#                             tf.where(tf.greater(y,y_low),tf.square(y-y),tf.square(y-y_low))  , 
#                             tf.square(y-y_up))
#                   )
#loss1 = tf.reduce_mean(tf.square(y-(y_up+y_low)/2))

#loss2 = tf.reduce_max(tf.square(y-y_up)+tf.square(y-y_low))

#loss = tf.reduce_mean(tf.square(y-y_))
#loss1 = tf.reduce_max(tf.square(y-y_))
#cross_entropy = tf.reduce_max(tf.square(y-y_))
#cross_entropy = tf.reduce_mean(tf.where(tf.greater(y_,y),  tf.square(y-y_), tf.square(y-y)))
#cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y_,logits=y))

#train_step=tf.train.AdamOptimizer(lr).minimize(loss)
#train_step2=tf.train.AdamOptimizer(lr).minimize(loss2)
#train_step=tf.train.Optimizer(0.01).minimize(cross_entropy)
#定义损失函数和反向传播算法
```
