```python
plt.ylim(0, 2.0)

with open('save/test_loss_unfeed_normal_3.csv') as csvfile_unfeed:
    reader = csv.reader(csvfile_unfeed)
    list_unfeed = []
    for row in reader:
        row = np.array(row)
        temp_list = []
        for i in range(len(row)):
            temp_list.append(float(row[i]))

        list_unfeed.append(temp_list)
    
    plt.plot(list_unfeed[0],list_unfeed[1],label='neural network')
    plt.ylabel('Testing Loss')
    plt.xlabel('Batch Epoch')


with open('save/test_loss_feed_normal_3.csv') as csvfile_feed:
    reader = csv.reader(csvfile_feed)
    list_feed = []
    for row in reader:
        row = np.array(row)
        temp_list = []
        for i in range(len(row)):
            temp_list.append(float(row[i]))

        list_feed.append(temp_list)
    plt.plot(list_feed[0],list_feed[1],label='feed distribution')
    plt.ylabel('Testing Expectation')
    plt.xlabel('Batch Epoch')
    #plt.title('Two or more lines on same plot with suitable legends ')
    # show a legend on the plot
    plt.legend()
    plt.show()
```
