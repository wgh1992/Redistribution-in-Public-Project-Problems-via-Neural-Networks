```python
def h_3_star(a, b, t):
    return a - min(a, t) + b - min(b, t) + max(min(a, t)+min(b, t), 2*t/3) + 1/2 * max(min(a, t)+min(b, t), t) - 1/2 * max(max(min(a, t), min(b, t)), 2*t/3) - t/6


def f_function(a, b, z):
    if(z >= 1):
        return (a+b)/2 + z/3
    else:
        return z/3 + h_3_star(a, b, 1-z)/2

def h_function(input_list):
    #input_list = sorted(input_list)
    g_list = []
    for j1 in range(len(input_list) ):
        for j2 in range(len(input_list)):
            if(j1 != j2):
                a = input_list[j1]
                b = input_list[j2]
                z = sum(input_list)- a-b

                g_list.append( f_function(a, b, z) * (Agent_number_n-1))
    h = sum(g_list) * 3 /  (Agent_number_n) /  (Agent_number_n-1) /  (Agent_number_n - 2)
    return h
                
                
x_list = []
y_list = []
z_list = []
result_list = []
training_supervised_label=[]
for index in range(len(training_data)):
    h_list = []
    for i in range(Agent_number_n):
        x_list.append(training_data[index][i][0])
        y_list.append(training_data[index][i][1])
        h = h_function(training_data[index][i])
        z_list.append(float(h))
        h_list.append(float(h))
    training_supervised_label.append(h_list)
    result_list.append(sum(h_list)/training_label[index]) 
    
    
print(max(result_list), min(result_list), max(result_list)-min(result_list))
```

```python
# coe , max/min , 0 0 0 , constant
def h_function(individual, data):
    res = 0
    for i in range(len(individual.h_function)):
        temp = individual.h_function[i]
        if(temp[1]==1):
            res += temp[0] * max(sum(temp[3]*data),temp[2])
        if(temp[1]==0):
            res += temp[0] * min(sum(temp[3]*data),temp[2])
    
    return res+individual.constant


class individual_class:
    def __init__(self):
        self.h_function = []
        self.constant = 0
        self.fitness = 0.0
    
    def Fitness(self):
        if(self.fitness==0 ):
            self.fitness = run_all_training_test(self)
        return self
    
    def debug(self):
        for i in range(len(self.h_function)):
            print("h_function: ",self.h_function[i])
        print("+",self.constant)
        print(self.fitness)
        print()
```

```python
def produce_initial_individual_list():
    length = random.randint(1,5)
    temp = individual_class()
    for i in range(length):
        temp.h_function.append(rand_sub_function())
    temp.constant = random.uniform(-1,1)
    return temp

global population,bid,t
population = []

def initialPopulation():
    global population,bid,t
    population = []
    for i in range(popSize):
        population.append(produce_initial_individual_list())
    return population

```

```python
def rank_heads(population):
    fitnessResults = []
    for i in range(0,len(population)):
        temp = population[i].Fitness()
        fitnessResults.append([i, temp.fitness, temp])
    fitnessResults = np.array(fitnessResults)
    
    return sorted(fitnessResults, key = operator.itemgetter(1))
```

```python
def selection(popRanked, eliteSize):
    selectionResults = []
    df = pd.DataFrame(np.array(popRanked), columns=["Index","Fitness","Tree"])
    df['cum_sum'] = df.Fitness.cumsum()
    df['cum_perc'] = 100*df.cum_sum/df.Fitness.sum()
    
    for i in range(int(len(popRanked))):
        selectionResults.append(popRanked[i][0])
    return selectionResults
```

```python
def matingPool(population, selectionResults):
    matingpool = []
    for i in range(0, len(selectionResults)):
        index = selectionResults[i]
        matingpool.append(population[index])
    return matingpool
```

```python
def breed(parent1, parent2):
    child = copy.deepcopy(parent1)
    
    geneA_1 = int(random.random() * len(parent1.h_function))
    
    geneA_2 = int(random.random() * len(parent2.h_function))
    
    child.h_function[geneA_1] = parent2.h_function[geneA_2]
    
    child.fitness = 0;
    return child

    
```

```python
def breedPopulation(matingpool, eliteSize):
    children = []
    length = len(matingpool) - eliteSize
    pool = random.sample(matingpool, len(matingpool))

    for i in range(eliteSize):
        children.append(matingpool[i])

    child = breed(pool[random.randint(0,len(matingpool)-1)], pool[random.randint(0,len(matingpool)-1)])
    if (child != False):

        child.fitness = 0
        children.append(child)


    return children

def mutate(individual, mutationRate):
    global NeighborhoodRate 
    
    temp_individual = copy.deepcopy(individual)

    for nerboorhood in range(len(temp_individual.h_function)):
        
#         if(random.random() < NeighborhoodRate):
#             if(random.random()>=0.5):
#                 temp_individual.h_function[nerboorhood][0]+= 0.1

#             else:
#                 temp_individual.h_function[nerboorhood][0]-= 0.1

#             temp_individual.fitness=0     
#         if(random.random() < NeighborhoodRate):
#             if(random.random()>=0.5):
#                 temp_individual.h_function[nerboorhood][2]+= 0.1

#             else:
#                 temp_individual.h_function[nerboorhood][2]-= 0.1

#             temp_individual.fitness=0     

        if(random.random() < NeighborhoodRate):
            del temp_individual.h_function[nerboorhood]

            temp_individual.fitness=0    
            break;
      
            
    if(temp_individual.Fitness().fitness<individual.Fitness().fitness):
        individual = temp_individual
    return individual

def mutatePopulation(population, mutationRate):
    mutatedPop = []

    for ind in range(0, len(population)):
        mutatedInd = mutate(copy.deepcopy(population[ind]), mutationRate)
        mutatedPop.append(mutatedInd)
    return mutatedPop
```

```python
def nextGeneration(currentGen, eliteSize, mutationRate):
    print(len(currentGen))
    popRanked = rank_heads(currentGen)

    selectionResults = selection(popRanked, eliteSize)
    
    matingpool = matingPool(currentGen, selectionResults)

    children = breedPopulation(matingpool, eliteSize)

    nextGeneration = mutatePopulation(children, mutationRate)

    for i in range(len(nextGeneration), popSize):
        nextGeneration.append( produce_initial_individual_list())

    return nextGeneration
```

```python
final_pop =[]
number_untruthful = 0
def geneticAlgorithm(pop, popSize, eliteSize, mutationRate, generations):

    print("Initial Fitness " + str(rank_heads(pop)[0][1]))
    best_Index = rank_heads(pop)[0][0]
    best_one = pop[best_Index]
    best_one.debug()
    progress = []
    progress.append(rank_heads(pop)[0][1])
    for i in range(0, generations):
        pop = nextGeneration(pop, eliteSize, mutationRate)
        print("Generation " , i , " : " + str(rank_heads(pop)[0][1]))
        progress.append(rank_heads(pop)[0][1])
        best_Index = rank_heads(pop)[0][0]
        best_one = pop[best_Index]
        best_one.debug()
        final_pop.append(pop[best_Index])
        print()
        
    for i in range (len(pop)):
        final_pop.append(pop[i])
    
    res_rank = rank_heads(final_pop)
    
    bestdeadlineIndex = res_rank[i][0]
    bestdeadline = final_pop[bestdeadlineIndex]

    print("Final Fitness: " + str(res_rank[i][1]))

    #print(progress)
    plt.plot(progress)
    plt.ylabel('Fitness')
    plt.xlabel('Generation')
    plt.show()
    return bestdeadline
```
